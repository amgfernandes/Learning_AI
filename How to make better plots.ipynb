```python
import numpy as np
import matplotlib.pyplot as plt
```

# based on https://www.youtube.com/watch?v=_56pLwmYJBs&feature=youtu.be&utm_medium=email&utm_source=getresponse&utm_content=Lesson+3%3A+Plot+like+a+designer&utm_campaign=

```python
labels=["milk", "dark", "hazelnut", "white", "orange"]
sales= np.array([80,83,63,66,49])
```

# Not a good plot

```python
fig, ax= plt.subplots(figsize=(6,6))
ax.set_axisbelow(True)
ax.yaxis.grid(color='k', linestyle='dashed')
ax.bar(labels, sales, label="Sales")
ax.legend(fancybox=True, shadow=True)
ax.set_ylim(20,100) #bad practice when comparing levels
ax.set_title("Units sold")
ax.set_xlabel("Type", fontsize=14)
ax.set_ylabel("Sales")
```

# Getting a better plot

```python
from matplotlib import cm
cmap= cm.get_cmap('magma')
loc =(sales-20)/(100-30)
colours=cmap(loc)

fig, ax= plt.subplots(figsize=(6,6))
ax.set_axisbelow(True)
ax.bar(labels, sales, label="Sales",color=colours)
for i,v in enumerate (sales):
    ax.text(i,v , v, ha="center", va="bottom",\
            fontsize=12, color="gray")
ax.set_ylabel("Sales", fontsize=14)
```

```python
from matplotlib import cm
cmap= cm.get_cmap('magma')
loc =(sales-30)/(100-30)
colours=cmap(loc)

fig, ax= plt.subplots(figsize=(6,6))
ax.set_axisbelow(True)
ax.bar(labels, sales, label="Sales",color=colours)
for i,v in enumerate (sales):
    ax.text(i,v , v, ha="center", va="bottom",\
            fontsize=12, color=colours[i])
ax.yaxis.set_ticks([])
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.text(2,-12, "Units sold of chocolate per week", fontsize=14, ha="center", alpha=0.7)
```
