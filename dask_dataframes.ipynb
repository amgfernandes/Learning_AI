# Based on https://examples.dask.org/dataframe.html

```python
import dask.dataframe as dd

import joblib
from distributed import Client, progress

client = Client(n_workers=4)
client
```

You can navigate to http://localhost:8787/status to see the diagnostic dashboard if you have Bokeh installed

```python
df=dd.read_csv('objects.csv', delimiter=';')
df
```

```python
df.shape
```

```python
df.head()
```

### Call .compute() when you want your result as a Pandas dataframe.

```python
df.groupby('ObjectNumber').Intensity_MeanIntensity_DNA.mean().compute()
```

```python
'''write to parquet'''
df.to_parquet('objects.parquet', engine='pyarrow')
```

```python
read_parquet = dd.read_parquet('objects.parquet/', engine='pyarrow')
read_parquet
```

```python
read_parquet.head()
```

# Groupby Aggregations

```python
df.groupby('ObjectNumber').agg({'Intensity_MeanIntensity_DNA': ['mean', 'std'], 'Intensity_MaxIntensity_DNA': ['mean', 'count']}).compute().head()
```

### Persist data in memory

If you have the available RAM for your dataset then you can persist data in memory.

This allows future computations to be much faster.

```python
df_ram = df.persist()
df_ram
```
